```python
city = ["北京市", "上海市", "天津市", "重庆市", "哈尔滨市", "长春市", "沈阳市", "呼和浩特市", "石家庄市"
        , "乌鲁木齐市", "兰州市", "西宁市", "西安市", "银川市", "郑州市", "济南市", "太原市", "合肥市", "武汉市",
        "长沙市", "南京市", "成都市", "贵阳市", "昆明市", "南宁市", "拉萨市","杭州市", "南昌市", "广州市","福州市",
        "台北市", "海口市", "香港", "澳门"]
len(city)
```

```python
city_location = []
for temp_city in city:
    print(temp_city)
    temp_loc = geolocator.geocode(temp_city)
    city_location.append((temp_loc.latitude, temp_loc.longitude))
city_location
```

```python
dis_info = {}
dis_info['位置'] = city
for i in range(len(city)):
    temp_dis_info = []
    start_geo = city_location[i]
    for end_loc in city_location:
        temp_dis_info.append(distance.distance(start_geo, end_loc, ellipsoid='GRS-80').km)
    dis_info[city[i]] = temp_dis_info
dis_info
```

```python
pd_dis = pd.DataFrame.from_dict(dis_info)
```
