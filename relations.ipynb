```python
def contain_user_info(data):
    return 'card_group' in data and len(data['card_group']) > 0 and 'user' in data['card_group'][0]
def find_user_index(card_info_list):
    index_list = []
    for i in range(len(card_info_list)-1, -1, -1):
        if contain_user_info(card_info_list[i]):
            index_list.append(i)
    return index_list
```

```python
def get_all_follow(id):
    id_list, name_list = [], []
    page = 1
    while(True):
        url = "http://m.weibo.cn/api/container/getIndex?containerid=231051_-_followers_-_{}&page={}".format(id, page)
        # print(url)
        data= requests.get(url,  headers=headers).json()
        if not data['ok']:
            break
        card_info_list = data['data']['cards']
        index_list = find_user_index(card_info_list)
        for index in index_list:
            user_info_list = card_info_list[index]['card_group']
            for temp_user_info in user_info_list:
                if "user" not in temp_user_info or temp_user_info["user"]==None:
                    continue
                else:
                    # if temp_user_info['user']['id'] not in id_list:
                    id_list.append(temp_user_info['user']['id'])
                    name_list.append(temp_user_info['user']['screen_name'])

        page += 1
        time.sleep(random() * per_sleep_time)

    return len(id_list), id_list, name_list

total_follow_num, follow_id_list, follow_name_list = get_all_follow(id)
print(total_follow_num)
for i in range(total_follow_num):
    print("{}----{}".format(follow_id_list[i], follow_name_list[i]))
```

```python
def get_co_user(fans_id_list, fans_name_list,follow_id_list ):
    co_id_list, co_name_list = [], []
    for i in range(len(fans_id_list)):
        if fans_id_list[i] in follow_id_list and fans_id_list[i] not in co_id_list:
            co_id_list.append(fans_id_list[i])
            co_name_list.append(fans_name_list[i])

    return co_id_list, co_name_list
co_id_list, co_name_list = get_co_user(fans_id_list, fans_name_list, follow_id_list)
print(len(co_id_list))
for i in range(len(co_id_list)):
    print("{}----{}".format(co_id_list[i], co_name_list[i]))
```

```python
#在获取用户关注，同时查看是否含有expected id
def test_userfollow_exist(id, expected_id):
    page = 1
    while(True):
        url = "http://m.weibo.cn/api/container/getIndex?containerid=231051_-_followers_-_{}&page={}".format(id, page)
        # print(url)
        data= requests.get(url,  headers=headers).json()
        if not data['ok']:
            break
        card_info_list = data['data']['cards']
        index = find_user_index(card_info_list)
        if index == -1:
            break
        user_info_list = card_info_list[index]['card_group']
        for temp_user_info in user_info_list:
            # if temp_user_info['user']['id'] not in id_list:
            if temp_user_info['user']['id'] == expected_id:
                return True

        page += 1
        time.sleep(random() * per_sleep_time)

    return False
```

```python
#基于用户关注列表，遍历每一个用户，查看他的关注列表是否含有原用户（较耗时）
def from_follow_to_cofollow(follow_id_list, follow_name_list, origin_id):
    cofollow_id_list, cofollow_name_list = [], []
    for i in tqdm(range(len(follow_id_list))):
        if test_userfollow_exist(follow_id_list[i], origin_id):
            cofollow_id_list.append(follow_id_list[i])
            cofollow_name_list.append(follow_name_list[i])
    return len(cofollow_id_list), cofollow_id_list, cofollow_name_list

co_count, co_id_list, co_name_list = from_follow_to_cofollow(follow_id_list, follow_name_list, id)

print(co_count)
for i in range(co_count):
    print("{}---{}".format(co_id_list[i], co_name_list[i]))
```
